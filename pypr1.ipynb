```python
print("welcome to the interactive personal data collector")

name = input("please enter your name :")
age = int(input("please enter your age :"))
height = float(input("please enter your height in meter :"))
number = int(input("enter your favourite number :"))

a = 2026 - age
halfage = age/2
n2 = number * 2

print("This program collects your personal information and demonstrates basic Python concepts like variables, data types, operators, and type casting.")
print("thank you! here is information we collected :")


print(f"name :{name}")
print("age was converted string into integer")
print(f"age :{age}")
print(f"half age :{halfage}")
print(f"double favourite number :{n2}")
print("height was converted string into float")
print(f"height :{height}")
print("number was converted string into integer")
print(f"favourite number :{number}")

print("name type :",type(name))
print("age type :",type(age))
print("height type :",type(height))
print("number type :",type(number))

print("name memory address :", id(name))
print("age memory address :", id(age))
print("height memory address :", id(height))
print("number memory address :", id(number))

print(f"your birth year is approximately :{a}")


print("thank you for using the personal Data collector!")
print("Keep learning and exploring Python to improve your programming skills. Goodbye !")


```
